### Nameraná prevodová charakteristika

```python
# ----------------------------------------------
# Zlucenie usekov do datasetu ustalenych stavov

dataRepoDir = './dataRepo/'

sch_files = [
    f for f in os.listdir(dataRepoDir)
    if f.startswith('SCH') and data_pot in f
]

steadyStateData = np.empty((0, 2))

for sch_file in sch_files:

    sch_data = np.loadtxt(
        dataRepoDir + sch_file,
        delimiter=',',
        skiprows=1,  
    )

    steadyStateData = np.vstack([steadyStateData, sch_data[:, 1:3]])

np.savetxt(dataRepoDir + 'ALLSCH_' + data_pot + '_steadyStateData' + '.csv',
    steadyStateData,
    delimiter=',',
    fmt='%.3f',
    header='sigOut,sigIn',
    comments='',
)


# ----------------------------------------------
# Vykreslenie dat ustalenych stavov

exec(open('./figjobs/figJob_03.py', encoding='utf-8').read())
```

```python
# ----------------------------------------------
# Identifikacia modelu na linearnom useku

tmpMask = (steadyStateData[:, 1] >= 0) & (steadyStateData[:, 1] <= 7.1)

identData = steadyStateData[tmpMask, :]

polyKoef = np.polyfit(
    identData[:, 1],
    identData[:, 0],
    deg=1
)



# plot data pre f_1
plot_u1 = np.arange(0, 7.1, 0.1)        
plot_y1 = np.polyval(polyKoef, plot_u1)

# plot data pre f_2
plot_u2 = np.arange(7, 10.1, 0.1)
plot_y2 = 10 * np.ones_like(plot_u2)

exec(open('./figjobs/figJob_04.py', encoding='utf-8').read())
```

### Aproximácia s využitím viacfaktorového modelu

```python

# ----------------------------------------------
# Dataset obsahujuci uplne vsetky ustalene stavy

allsch_files = [
        f for f in os.listdir(dataRepoDir)
        if f.startswith('ALLSCH') 
    ]

allDteadyStateData = np.empty((0, 3))

for allsch_file in allsch_files:
    allsch_data = np.loadtxt(
        dataRepoDir + allsch_file,
        delimiter=',',
        skiprows=1,  
    )

    allDteadyStateData = np.vstack([allDteadyStateData, allsch_data])



tmp_mask = np.logical_and(
    allDteadyStateData[:, 0] >= 0.1, 
    allDteadyStateData[:, 0] <= 9.9
)    


vector_y = allDteadyStateData[tmp_mask, 0]

matrix_H = np.column_stack((
    allDteadyStateData[tmp_mask, 1],
    allDteadyStateData[tmp_mask, 2],
    np.ones_like(allDteadyStateData[tmp_mask, 1])
))

# Vypocet parametrov modelu
theta = np.linalg.lstsq(matrix_H, vector_y, rcond=None)[0]
print('Parametre modelu: ', theta)


plot_mesh = np.meshgrid(
    np.arange(0, 10.2, 0.2),
    np.arange(0, 10.2, 0.2)
)

plot_y1 = theta[0] * plot_mesh[0] + theta[1] * plot_mesh[1] + theta[2]



# ----------------------------------------------
# 3D vykreslenie dat ustalenych stavov

exec(open('./figjobs/figJob_07.py', encoding='utf-8').read())

```
