```python
# Obr. exp3 prevodova charakteristika with linear fit and saturation

from matplotlib.ticker import MultipleLocator, AutoMinorLocator

figName = 'exp3_prevodova_fit_sat'

# ===== Load steady-state points =====
exp_folder = os.path.join(processed_dir, 'exp3')

data_sp = pd.read_csv(
    os.path.join(exp_folder, 'exp3_steady_points.csv')
)

u_sp = data_sp.iloc[:, sp_u].values
y1_sp = data_sp.iloc[:, sp_y1].values
y2_sp = data_sp.iloc[:, sp_y2].values

# ===== Masks =====
mask_lin = np.logical_and(u_sp >= 0, u_sp <= 7)
mask_sat = np.logical_and(u_sp >= 7, u_sp <= 10)

u_lin = u_sp[mask_lin]
y1_lin = y1_sp[mask_lin]
y2_lin = y2_sp[mask_lin]

u_sat = u_sp[mask_sat]
y1_sat = y1_sp[mask_sat]
y2_sat = y2_sp[mask_sat]

# ===== Polyfit for linear region =====
coef_y1 = np.polyfit(u_lin, y1_lin, 1)
coef_y2 = np.polyfit(u_lin, y2_lin, 1)

u_fit_lin = np.linspace(0, 7, 200)

y1_fit_lin = np.polyval(coef_y1, u_fit_lin)
y2_fit_lin = np.polyval(coef_y2, u_fit_lin)

# ===== Saturation =====
y1_sat_mean = np.mean(y1_sat)
y2_sat_mean = np.mean(y2_sat)

u_fit_sat = np.linspace(7, 10, 100)

y1_fit_sat = np.full_like(u_fit_sat, y1_sat_mean)
y2_fit_sat = np.full_like(u_fit_sat, y2_sat_mean)

# ===== Figure =====
figPlotParam = fcnDefaultFigSize(5.0, 0.15, 1-0.14, 0.14, 0.7, 13)

fig = plt.figure(0, figsize=figPlotParam[0:2])
ax = plt.subplot(111)

# ===== Measured points =====
ax.plot(u_sp, y1_sp, 'o',
        ms=6,
        mfc='none',
        mec='red',
        mew=0.1,
        linestyle='None',
        label=r'$y_1$ merané')

ax.plot(u_sp, y2_sp, 'o',
        ms=6,
        mfc='none',
        mec='green',
        mew=0.1,
        linestyle='None',
        label=r'$y_2$ merané')

# ===== Model y1 =====
ax.plot(u_fit_lin, y1_fit_lin, '--',
        lw=1.0,
        color='k',
        label=r'$y_1$ model')

ax.plot(u_fit_sat, y1_fit_sat, '--',
        lw=1.0,
        color='k')

# ===== Model y2 =====
ax.plot(u_fit_lin, y2_fit_lin, '--',
        lw=1.0,
        color='grey',
        label=r'$y_2$ model')

ax.plot(u_fit_sat, y2_fit_sat, '--',
        lw=1.0,
        color='grey')

# ===== Labels =====
ax.set_xlabel(r'$u$ [V]', ha='left', va='top')
ax.xaxis.set_label_coords(1.05, -0.07, transform=ax.transAxes)

ax.set_ylabel('[V]', ha='right', va='bottom', rotation=0)
ax.yaxis.set_label_coords(-0.07, 1.05, transform=ax.transAxes)

ax.set_title('Prevodová charakteristika s aproximáciou a saturáciou',
             ha='left', va='bottom', x=0.0, y=1.02)

# ===== Grid =====
ax.xaxis.set_major_locator(MaxNLocator(nbins=11))
ax.xaxis.set_minor_locator(AutoMinorLocator())

ax.yaxis.set_major_locator(MaxNLocator(nbins=8))
ax.yaxis.set_minor_locator(AutoMinorLocator())

fcnDefaultAxisStyle(ax)

# ===== Legend =====
ax.legend(frameon=False,
          fontsize=9,
          loc='upper right',
          bbox_to_anchor=(1.2, 1.0))

# ===== Layout =====
fcnDefaultLayoutAdj(fig,
                    figPlotParam[2],
                    figPlotParam[3],
                    figPlotParam[4],
                    figPlotParam[5])

# ===== Save =====
plt.savefig(os.path.join(fig_save_dir, f"{figName}.png"), dpi=600)
plt.close()

print(f"Saved {figName}.png")

print("y1 linear fit coefficients [a, b]:", coef_y1)
print("y2 linear fit coefficients [a, b]:", coef_y2)

print(f"y1 linear part: y = {coef_y1[0]:.6f} * u + {coef_y1[1]:.6f}")
print(f"y2 linear part: y = {coef_y2[0]:.6f} * u + {coef_y2[1]:.6f}")

print(f"y1 saturation part: y = {y1_sat_mean:.6f}")
print(f"y2 saturation part: y = {y2_sat_mean:.6f}")
```

```python
# Obr. 3D static characteristic with linear plane model

from mpl_toolkits.mplot3d import Axes3D

figName = 'allExp_3D_static_characteristic_with_plane'

# ===== Collect data from all experiments =====
x_all_black = []   # u2
y_all_black = []   # u
z1_all_black = []  # y1
z2_all_black = []  # y2

x_all_red = []
y_all_red = []
z1_all_red = []
z2_all_red = []

x_exp1_cross = []
y_exp1_cross = []

experiment_names = ['exp1', 'exp2', 'exp3', 'exp4', 'exp5']

for exp_name in experiment_names:

    exp_folder = os.path.join(processed_dir, exp_name)

    data_sp = pd.read_csv(
        os.path.join(exp_folder, f'{exp_name}_steady_points.csv')
    )

    u = data_sp.iloc[:, sp_u].values
    u2 = data_sp.iloc[:, sp_u2].values
    y1 = data_sp.iloc[:, sp_y1].values
    y2 = data_sp.iloc[:, sp_y2].values

    # Normal region
    if exp_name == 'exp1':
        mask_black = u < 7
    else:
        mask_black = u <= 7

    x_all_black.extend(u2[mask_black])
    y_all_black.extend(u[mask_black])
    z1_all_black.extend(y1[mask_black])
    z2_all_black.extend(y2[mask_black])

    # Saturation / high-input region
    if exp_name == 'exp1':
        mask_red = u >= 7
    else:
        mask_red = u > 7

    if exp_name == 'exp1':
        x_exp1_cross.extend(u2[mask_red])
        y_exp1_cross.extend(u[mask_red])
    else:
        x_all_red.extend(u2[mask_red])
        y_all_red.extend(u[mask_red])
        z1_all_red.extend(y1[mask_red])
        z2_all_red.extend(y2[mask_red])

# ===== Convert to numpy arrays =====
x_all_black = np.array(x_all_black)
y_all_black = np.array(y_all_black)
z1_all_black = np.array(z1_all_black)
z2_all_black = np.array(z2_all_black)

x_all_red = np.array(x_all_red)
y_all_red = np.array(y_all_red)
z1_all_red = np.array(z1_all_red)
z2_all_red = np.array(z2_all_red)

x_exp1_cross = np.array(x_exp1_cross)
y_exp1_cross = np.array(y_exp1_cross)

z1_exp1_cross = np.full_like(x_exp1_cross, 10.0, dtype=float)
z2_exp1_cross = np.full_like(x_exp1_cross, 8.0, dtype=float)

# ===== Linear model of two variables =====
# Model form:
# y = a*u2 + b*u + c

H = np.column_stack((
    x_all_black,   # u2
    y_all_black,   # u
    np.ones_like(x_all_black)
))

theta_y1, _, _, _ = np.linalg.lstsq(H, z1_all_black, rcond=None)
theta_y2, _, _, _ = np.linalg.lstsq(H, z2_all_black, rcond=None)

# ===== Grid for plane =====
u2_grid = np.linspace(0, 10, 30)
u_grid = np.linspace(0, 10, 30)

U2_grid, U_grid = np.meshgrid(u2_grid, u_grid)

Y1_plane = theta_y1[0] * U2_grid + theta_y1[1] * U_grid + theta_y1[2]
Y2_plane = theta_y2[0] * U2_grid + theta_y2[1] * U_grid + theta_y2[2]

Y1_plane[(Y1_plane < 0) | (Y1_plane > 10)] = np.nan
Y2_plane[(Y2_plane < 0) | (Y2_plane > 10)] = np.nan

# ===== Figure =====
figPlotParam = fcnDefaultFigSize(9, 0.08, 1-0.1, 0.1, 0.0, 16)

fig = plt.figure(0, figsize=figPlotParam[0:2])

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax2 = fig.add_subplot(1, 2, 2, projection='3d')

# ===== Left panel: y1 =====
ax1.scatter(x_all_black, y_all_black, z1_all_black,
            c='k', s=2, marker='o', depthshade=False,
            label=r'$y_1$, $u \leq 7$')

ax1.scatter(x_all_red, y_all_red, z1_all_red,
            c='red', s=2, marker='o', depthshade=False,
            label=r'$y_1$, $u > 7$')

ax1.scatter(x_exp1_cross, y_exp1_cross, z1_exp1_cross,
            c='red', s=8, marker='x', depthshade=False,
            label=r'$y_1$, p.p. 1 havarijný režim')

ax1.plot_wireframe(U2_grid, U_grid, Y1_plane,
                   rstride=1, cstride=1,
                   color='purple', linewidth=0.12)

ax1.set_xlabel(r'Prevádzková podmienka [V]', labelpad=6)
ax1.set_ylabel(r'Vstupný signál [V]', labelpad=6)
ax1.set_zlabel(r'Výstupný signál [V]', labelpad=6)

ax1.set_title(r'$y_1$', loc='left', y=0.98)

ax1.set_xlim(0, 10)
ax1.set_ylim(0, 10)
ax1.set_zlim(0, 10)

ax1.view_init(elev=12, azim=-55)

# ===== Right panel: y2 =====
ax2.scatter(x_all_black, y_all_black, z2_all_black,
            c='k', s=2, marker='o', depthshade=False,
            label=r'$y_2$, $u \leq 7$')

ax2.scatter(x_all_red, y_all_red, z2_all_red,
            c='red', s=2, marker='o', depthshade=False,
            label=r'$y_2$, $u > 7$')

ax2.scatter(x_exp1_cross, y_exp1_cross, z2_exp1_cross,
            c='red', s=8, marker='x', depthshade=False,
            label=r'$y_2$, p.p. 1 havarijný režim')

ax2.plot_wireframe(U2_grid, U_grid, Y2_plane,
                   rstride=1, cstride=1,
                   color='purple', linewidth=0.12)

ax2.set_xlabel(r'Prevádzková podmienka [V]', labelpad=6)
ax2.set_ylabel(r'Vstupný signál [V]', labelpad=6)
ax2.set_zlabel(r'Výstupný signál [V]', labelpad=6)

ax2.set_title(r'$y_2$', loc='left', y=0.98)

ax2.set_xlim(0, 10)
ax2.set_ylim(0, 10)
ax2.set_zlim(0, 10)

ax2.view_init(elev=12, azim=-55)

# ===== Legends =====
ax1.legend(frameon=False, fontsize=10, loc='upper left', bbox_to_anchor=(0.00, 1.00))
ax2.legend(frameon=False, fontsize=10, loc='upper left', bbox_to_anchor=(0.00, 1.00))

# ===== Layout =====
fcnDefaultLayoutAdj(fig,
                    figPlotParam[2],
                    figPlotParam[3],
                    figPlotParam[4],
                    figPlotParam[5])

fig.subplots_adjust(wspace=0.25)

# ===== Save =====
plt.savefig(os.path.join(fig_save_dir, f"{figName}.png"), dpi=600)
plt.close()

# ===== Print results =====
print(f"Saved {figName}.png")
print("Used experiments:", experiment_names)

print("\nLinear model for y1:")
print("theta =", theta_y1)
print(f"y1 = {theta_y1[0]:.6f} * u2 + {theta_y1[1]:.6f} * u + {theta_y1[2]:.6f}")

print("\nLinear model for y2:")
print("theta =", theta_y2)
print(f"y2 = {theta_y2[0]:.6f} * u2 + {theta_y2[1]:.6f} * u + {theta_y2[2]:.6f}")
```

```python
# Obr. 3D static characteristic with quadratic surface model

from mpl_toolkits.mplot3d import Axes3D

figName = 'allExp_3D_static_characteristic_with_quad_surface'

# ===== Collect data from all experiments =====
x_all_black = []   # u2
y_all_black = []   # u
z1_all_black = []  # y1
z2_all_black = []  # y2

x_all_red = []
y_all_red = []
z1_all_red = []
z2_all_red = []

x_exp1_cross = []
y_exp1_cross = []

experiment_names = ['exp1', 'exp2', 'exp3', 'exp4', 'exp5']

for exp_name in experiment_names:

    exp_folder = os.path.join(processed_dir, exp_name)

    data_sp = pd.read_csv(
        os.path.join(exp_folder, f'{exp_name}_steady_points.csv')
    )

    u = data_sp.iloc[:, sp_u].values
    u2 = data_sp.iloc[:, sp_u2].values
    y1 = data_sp.iloc[:, sp_y1].values
    y2 = data_sp.iloc[:, sp_y2].values

    # Normal region
    if exp_name == 'exp1':
        mask_black = u < 7
    else:
        mask_black = u <= 7

    x_all_black.extend(u2[mask_black])
    y_all_black.extend(u[mask_black])
    z1_all_black.extend(y1[mask_black])
    z2_all_black.extend(y2[mask_black])

    # Saturation / high-input region
    if exp_name == 'exp1':
        mask_red = u >= 7
    else:
        mask_red = u > 7

    if exp_name == 'exp1':
        x_exp1_cross.extend(u2[mask_red])
        y_exp1_cross.extend(u[mask_red])
    else:
        x_all_red.extend(u2[mask_red])
        y_all_red.extend(u[mask_red])
        z1_all_red.extend(y1[mask_red])
        z2_all_red.extend(y2[mask_red])

# ===== Convert to numpy arrays =====
x_all_black = np.array(x_all_black)
y_all_black = np.array(y_all_black)
z1_all_black = np.array(z1_all_black)
z2_all_black = np.array(z2_all_black)

x_all_red = np.array(x_all_red)
y_all_red = np.array(y_all_red)
z1_all_red = np.array(z1_all_red)
z2_all_red = np.array(z2_all_red)

x_exp1_cross = np.array(x_exp1_cross)
y_exp1_cross = np.array(y_exp1_cross)

z1_exp1_cross = np.full_like(x_exp1_cross, 10.0, dtype=float)
z2_exp1_cross = np.full_like(x_exp1_cross, 8.0, dtype=float)

# ===== Quadratic model of two variables =====
# Model form:
# y = a*u2^2 + b*u^2 + c*u2*u + d*u2 + e*u + f

H = np.column_stack((
    x_all_black**2,                 # u2^2
    y_all_black**2,                 # u^2
    x_all_black * y_all_black,      # u2*u
    x_all_black,                    # u2
    y_all_black,                    # u
    np.ones_like(x_all_black)       # 1
))

theta_y1, _, _, _ = np.linalg.lstsq(H, z1_all_black, rcond=None)
theta_y2, _, _, _ = np.linalg.lstsq(H, z2_all_black, rcond=None)

# ===== Grid for surface =====
u2_grid = np.linspace(0, 10, 30)
u_grid = np.linspace(0, 10, 30)

U2_grid, U_grid = np.meshgrid(u2_grid, u_grid)

Y1_surface = (
    theta_y1[0] * U2_grid**2 +
    theta_y1[1] * U_grid**2 +
    theta_y1[2] * U2_grid * U_grid +
    theta_y1[3] * U2_grid +
    theta_y1[4] * U_grid +
    theta_y1[5]
)

Y2_surface = (
    theta_y2[0] * U2_grid**2 +
    theta_y2[1] * U_grid**2 +
    theta_y2[2] * U2_grid * U_grid +
    theta_y2[3] * U2_grid +
    theta_y2[4] * U_grid +
    theta_y2[5]
)

# Do not draw outside physical range
Y1_surface[(Y1_surface < 0) | (Y1_surface > 10)] = np.nan
Y2_surface[(Y2_surface < 0) | (Y2_surface > 10)] = np.nan

# ===== Figure =====
figPlotParam = fcnDefaultFigSize(9, 0.08, 1-0.1, 0.1, 0.0, 16)

fig = plt.figure(0, figsize=figPlotParam[0:2])

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax2 = fig.add_subplot(1, 2, 2, projection='3d')

# ===== Left panel: y1 =====
ax1.scatter(x_all_black, y_all_black, z1_all_black,
            c='k', s=2, marker='o', depthshade=False,
            label=r'$y_1$, $u \leq 7$')

ax1.scatter(x_all_red, y_all_red, z1_all_red,
            c='red', s=2, marker='o', depthshade=False,
            label=r'$y_1$, $u > 7$')

ax1.scatter(x_exp1_cross, y_exp1_cross, z1_exp1_cross,
            c='red', s=8, marker='x', depthshade=False,
            label=r'$y_1$, p.p. 1 havarijný režim')

ax1.plot_wireframe(U2_grid, U_grid, Y1_surface,
                   rstride=1, cstride=1,
                   color='purple', linewidth=0.12)

ax1.set_xlabel(r'Prevádzková podmienka [V]', labelpad=6)
ax1.set_ylabel(r'Vstupný signál [V]', labelpad=6)
ax1.set_zlabel(r'Výstupný signál [V]', labelpad=6)

ax1.set_title(r'$y_1$', loc='left', y=0.98)

ax1.set_xlim(0, 10)
ax1.set_ylim(0, 10)
ax1.set_zlim(0, 10)

ax1.view_init(elev=12, azim=-55)

# ===== Right panel: y2 =====
ax2.scatter(x_all_black, y_all_black, z2_all_black,
            c='k', s=2, marker='o', depthshade=False,
            label=r'$y_2$, $u \leq 7$')

ax2.scatter(x_all_red, y_all_red, z2_all_red,
            c='red', s=2, marker='o', depthshade=False,
            label=r'$y_2$, $u > 7$')

ax2.scatter(x_exp1_cross, y_exp1_cross, z2_exp1_cross,
            c='red', s=8, marker='x', depthshade=False,
            label=r'$y_2$, p.p. 1 havarijný režim')

ax2.plot_wireframe(U2_grid, U_grid, Y2_surface,
                   rstride=1, cstride=1,
                   color='purple', linewidth=0.12)

ax2.set_xlabel(r'Prevádzková podmienka [V]', labelpad=6)
ax2.set_ylabel(r'Vstupný signál [V]', labelpad=6)
ax2.set_zlabel(r'Výstupný signál [V]', labelpad=6)

ax2.set_title(r'$y_2$', loc='left', y=0.98)

ax2.set_xlim(0, 10)
ax2.set_ylim(0, 10)
ax2.set_zlim(0, 10)

ax2.view_init(elev=12, azim=-55)

# ===== Legends =====
ax1.legend(frameon=False, fontsize=10, loc='upper left', bbox_to_anchor=(0.00, 1.00))
ax2.legend(frameon=False, fontsize=10, loc='upper left', bbox_to_anchor=(0.00, 1.00))

# ===== Layout =====
fcnDefaultLayoutAdj(fig,
                    figPlotParam[2],
                    figPlotParam[3],
                    figPlotParam[4],
                    figPlotParam[5])

fig.subplots_adjust(wspace=0.25)

# ===== Save =====
plt.savefig(os.path.join(fig_save_dir, f"{figName}.png"), dpi=600)
plt.close()

# ===== Print results =====
print(f"Saved {figName}.png")
print("Used experiments:", experiment_names)

print("\nQuadratic model for y1:")
print("theta =", theta_y1)
print(
    f"y1 = {theta_y1[0]:.6f} * u2^2 + "
    f"{theta_y1[1]:.6f} * u^2 + "
    f"{theta_y1[2]:.6f} * u2*u + "
    f"{theta_y1[3]:.6f} * u2 + "
    f"{theta_y1[4]:.6f} * u + "
    f"{theta_y1[5]:.6f}"
)

print("\nQuadratic model for y2:")
print("theta =", theta_y2)
print(
    f"y2 = {theta_y2[0]:.6f} * u2^2 + "
    f"{theta_y2[1]:.6f} * u^2 + "
    f"{theta_y2[2]:.6f} * u2*u + "
    f"{theta_y2[3]:.6f} * u2 + "
    f"{theta_y2[4]:.6f} * u + "
    f"{theta_y2[5]:.6f}"
)
```
